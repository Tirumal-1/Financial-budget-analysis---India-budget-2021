### I can see that the finance department is getting the most of the share from the total budget of the goverment. now let's plot this data into a dount plot to have a clear view of the distribution of funds among all the departments:

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]

plt.figure(figsize = (7,7))
plt.pie(df, labels = labels, autopct = "1%.1f%%", startangle = 90, pctdistance = 0.85, shadow = True)
central_circle = plt.Circle((0,0),0.5, color = "white")

fig  = plt.gcf()

fig.gca().add_artist(central_circle)
plt.rc("font", size = 12)
plt.title("Distribution of The Budget", fontsize = 20)
plt.show()
```

```python
# Ensure X is a DataFrame
X = data[['Department /Ministry']]  # Double square brackets to get a DataFrame

# Ensure Y is a DataFrame (if required by your model, usually it should be a Series)
Y = data['Fund allotted(in ₹crores)']

# Splitting the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Fitting the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
Y_pred=lr.predict(X_test)
    #find the mean squared error
print("Mean Squared error : ",mean_squared_error(Y_test,Y_pred))
    #find the r2_score 
print("R2 Score : ",r2_score(Y_test,Y_pred))
```

```python
#Given to Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(max_depth=3)
```

```python
rfr.fit(X_train, Y_train)
```

```python
Y_pred1=rfr.predict(X_test)
    #find the mean squared error
print("Mean Squared error : ",mean_squared_error(Y_test,Y_pred1))
    #find the r2_score 
print("R2 Score : ",r2_score(Y_test,Y_pred1))
```
